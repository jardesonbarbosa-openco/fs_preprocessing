```python
import numpy as np
import pandas as pd
```

```python
path = '../data/Dicionario_de_Agencias_e_Postos_Bancarios_com_Apelidos_v28.sav'
branch_df = pd.read_spss(path)
```

```python
def get_branch_code(bank, category):
    
    bank_high_income = {
        '341': 'PERS',
        '001': 'STIL',
        '237': 'PRIM'
    }

    bank_prime_location = {
        '399': 'HSBC',
        '033': 'VANG',
        '341': 'UNIC',
        '001': 'ESPA'
    }

    if category == 'Alta Renda':
        return bank_high_income.get(bank, 'OUTR')
    elif category == 'Localização Privilegiada':
        return bank_prime_location.get(bank, 'PRIV')

def map_branch_code(df: pd.DataFrame, bank_col:str, bank_cat_col:str):
    arr = df[[bank_col, bank_cat_col]].values

    vfunc = np.vectorize(get_branch_code)
    return vfunc(arr[:,0], arr[:,1])
```

```python
branch_df['branch_code'] = map_branch_code(branch_df, 'Bank', 'BankCategory')
branch_df['branch_code'] = branch_df['branch_code'].fillna(np.nan)
```

```python
branch_df.to_parquet('../data/bank_branch.parquet')
```
